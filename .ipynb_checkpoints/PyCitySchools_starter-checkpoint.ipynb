```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school = school_data_complete["school_name"].unique()
total_schools = len(school)
total_schools

total_student = len(school_data_complete)

district_real_student_count = student_data['student_name'].count()



list_budget = school_data_complete.groupby("school_name").mean()
total_budget = list_budget["budget"].sum()
#total_budget = '$[:,.2f]'.format(total_budget) didn't work for some reason
#total_budget = '$[:,]'.format(total_budget) neither did this, both printed wrong things
total_budget = '${:,.2f}'.format(total_budget)


average_math_score = school_data_complete['math_score'].mean()
average_reading_score = school_data_complete['reading_score'].mean()

#passing grades  

passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70,:]
reading_percentage = passing_reading['reading_score'].count()/ total_student * 100
reading_percentage

passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70,:]
math_percentage = passing_math['math_score'].count()/ total_student * 100
math_percentage

total_passing = (reading_percentage + math_percentage)/district_real_student_count
#total_passing = student_data[(student_data['math_score']>=70)&(student_data['reading_score']>=70)]['student_name'].count()/district_real_student_count
#total_passing = (int(passing_reading + passing_math)/2)

total_student = '{:,.0f}'.format(total_student)

district_summaryDF = pd.DataFrame({"Total Schools" : [total_schools, 1], "Total Students" : [total_student, 1],
                         "Total Budget": [total_budget, 1],
                           "Average Math Score" : [average_math_score, 1], "Average Reading Score" : [average_reading_score, 1],
                          "% Passing Math" : [math_percentage, 1], "% Passing Reading" : [reading_percentage,1], 
                          "% Overall Passing Rate" : [total_passing,1]})

summary = district_summaryDF.drop([1])

summary


```
